
Data ExplorationFirst, we'll load the Titanic dataset and briefly showcase it using pandas.

```python
import pandas as pd
df = pd.read_csv("./data_sets/Titanic-Dataset.csv", index_col="PassengerId")
df.head(5)
df.dropna()
```

```python
df.info()
```

Numerical columns: Age, Fare, SibSp, Parch

Categorical: Survived, Pclass, Sex, Embarked

Demonstrate creating a new feature, like "FamilySize," by combining SibSp and Parch. 

```python
df["FamilySize"]= df["SibSp"] + df["Parch"]
```

```python
dummies = pd.get_dummies(df["Sex"])
df.drop(["Sex"], axis=1, inplace=True)
```

```python
df= df.join(dummies)
df.head()
```

```python
df["Age"] = df["Age"].fillna(0.0).astype(int)
```
